```python
sns.barplot(data['Q3'])
plt.ylabel('Gender')
plt.xlabel('Count')
plt.grid(True)
plt.show()
```

```python
demographic_column='Q3'
top_countries=data['Q4'].value_counts().head(5).index
filtered_data=data[data['Q4'].isin(top_countries)]
sns.countplot(x=demographic_column,hue='Q4',data=filtered_data)
plt.title(f'Top 5 countries of Respondents and their gender')
plt.xlabel(demographic_column)
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
#research_published_count=data['Q9'].value_counts()
sns.countplot(data['Q9'])
plt.ylabel('No. of respondants Published Research')

plt.show()
```

```python
sns.countplot(x='Q11',hue='Q3',data=data)
plt.title('Coding experience of Respondents Based on gender')
plt.xlabel('Coding experience')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
sns.countplot(x='Q10_1',hue='Q3', data=data)
plt.title('Machine learning platform usage by gender')
plt.xlabel('ML usage')
plt.ylabel('Count')
plt.show()
```

```python
heatmap_data=pd.crosstab(index=data['Q23'],columns=[data['Q3']],margins=True, margins_name='Total')
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu',fmt='g',cbar_kws={'label':'Count'})
plt.show()
```
