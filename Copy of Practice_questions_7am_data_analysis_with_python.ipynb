# Week 3 Assignment : Matplotlib and Seaborn

1. Create a histogram of passenger ages.
2. Create a scatter plot of age vs. fare.
3. Analyze the dataset and plot comparing survival rates by passenger class.
4. Show the fare distribution for each passenger class.
5. Create a violin plot showing passenger age distributions for each passenger class.
6. Analyze the given data and visualize relationships between multiple numerical variables


````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
data=sns.load_dataset('titanic')
data.head()
````

````python
#Q NO.1:
plt.hist(data["age"],color="purple",alpha=0.6)
plt.title('Histogram of passenger ages')
plt.xlabel("Ages of passenger")
plt.ylabel('Frequency')
plt.show()
````

````python
#Q No 2:
x=data['age']
y=data['fare']
plt.scatter(x,y)
plt.title('Scatterplot of Age vs Fare ')
plt.xlabel("Age")
plt.ylabel('Fare')
plt.show()
````

````python
#Q No 3:
sns.barplot(x=data.pclass,y=data.survived , color='brown',alpha=0.7)
plt.title('Survival rate by passenger class')
plt.xlabel("Passenger class")
plt.ylabel('Survival rate')
plt.show()
````

````python
#Q no 4:
sns.boxplot(x=data.pclass,y=data.fare)
plt.title('Fare distribution for each passenger class')
plt.show()
````

````python
#Q no.5:
sns.violinplot(data=data,x='pclass',y='age')
plt.show()
````

````python
#Q no.6:
sns.heatmap(data[['age','fare','sibsp','parch']].corr())
plt.show()
````
